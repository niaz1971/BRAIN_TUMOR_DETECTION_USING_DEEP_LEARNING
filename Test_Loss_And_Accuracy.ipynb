```python
import matplotlib.pyplot as plt
import numpy as np

models = [
    "VGG16", "VGG19", "Inception_V3", "Xception", "ResNet101",
    "EfficientNetB0", "InceptionResNetV2", "MobileNetV2",
    "DenseNet121", "MobileNet1"
]

test_loss = [
    0.15928, 0.09114, 0.20791, 0.14486, 0.03686,
    1.28695, 0.18764, 0.12252, 0.18198, 0.08997
]

test_accuracy = [
    0.95652, 0.97559, 0.92067, 0.97407, 0.99008,
    0.49047, 0.94966, 0.96720, 0.95957, 0.97941
]

test_accuracy_percent = [a * 100 for a in test_accuracy]

x = np.arange(len(models))
width = 0.4

fig, ax1 = plt.subplots(figsize=(12, 6))

bars1 = ax1.bar(x - width/2, test_loss, width, label='Test Loss', color='salmon')
ax1.set_ylabel('Test Loss')
ax1.set_xticks(x)
ax1.set_xticklabels(models, rotation=45, ha='right')

ax2 = ax1.twinx()
bars2 = ax2.bar(x + width/2, test_accuracy_percent, width, label='Test Accuracy (%)', color='skyblue')
ax2.set_ylabel('Test Accuracy (%)')

plt.title('Transfer Learning Models: Test Loss and Test Accuracy (%)')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
