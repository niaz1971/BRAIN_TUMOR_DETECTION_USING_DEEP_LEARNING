```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- Setup Data ---
data = {'Label': ['notumor', 'pituitary', 'meningioma', 'glioma'],
        'Count': [1596, 1457, 1530, 1321]}
df = pd.DataFrame(data)

# Create the categorical column for plotting
train_data = pd.DataFrame({
    'Label': df['Label'].repeat(df['Count'])
})
# ----------------------------------------------------------

# Set a different style: 'dark' background
sns.set_style("dark")

plt.figure(figsize=(8, 5))

# Use a different color palette: 'Set1'
ax = sns.countplot(
    data=train_data,
    x='Label',
    order=train_data['Label'].value_counts().index, # Retain sorting by count
    palette='Set1'
)

# Annotate bars with the exact count
for p in ax.patches:
    ax.annotate(
        format(p.get_height(), '.0f'),
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha = 'center',
        va = 'center',
        xytext = (0, 9),
        textcoords = 'offset points',
        fontsize=10,
        fontweight='bold'
    )

plt.title('Distribution of Training Data Classes', fontsize=14)
plt.xlabel('Tumor Type', fontsize=12)
plt.ylabel('Count (N)', fontsize=12)
plt.tight_layout()
plt.show()
```
