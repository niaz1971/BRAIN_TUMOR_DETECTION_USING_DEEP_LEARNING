```python
import os
import random
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
import cv2
import imutils
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.models import Model
from tensorflow.keras.layers import GlobalAveragePooling2D, Dropout, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Clone the repo if not already done
repo_url = "https://github.com/nazianafis/Brain-MRI-Classification.git"
clone_dir = "./Brain-MRI-Classification"
if not os.path.exists(clone_dir):
    os.system(f"git clone {repo_url}")
else:
    print(f"Repository already cloned at {clone_dir}")

# Set dataset paths based on cloned repo
original_train_dir = os.path.join(clone_dir, "Brain-MRI", "Training")
original_test_dir = os.path.join(clone_dir, "Brain-MRI", "Testing")

if not os.path.exists(original_train_dir) or not os.path.exists(original_test_dir):
    raise FileNotFoundError("Training or Testing directories are missing in the cloned repo path.")

classes = ['glioma_tumor', 'meningioma_tumor', 'no_tumor', 'pituitary_tumor']

# Prepare folders for cropped images
base_train_dir = "./Crop-Brain-MRI"
base_test_dir = "./Test-Brain-MRI"
for base_dir in [base_train_dir, base_test_dir]:
    os.makedirs(base_dir, exist_ok=True)
    for cls in classes:
        os.makedirs(os.path.join(base_dir, cls), exist_ok=True)

def crop_image(image, plot=False):
    img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    img_gray = cv2.GaussianBlur(img_gray, (5, 5), 0)
    _, img_thresh = cv2.threshold(img_gray, 45, 255, cv2.THRESH_BINARY)
    img_thresh = cv2.erode(img_thresh, None, iterations=2)
    img_thresh = cv2.dilate(img_thresh, None, iterations=2)
    contours = cv2.findContours(img_thresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    contours = imutils.grab_contours(contours)

    if contours:
        c = max(contours, key=cv2.contourArea)
        extLeft = tuple(c[c[:, :, 0].argmin()][0])
        extRight = tuple(c[c[:, :, 0].argmax()][0])
        extTop = tuple(c[c[:, :, 1].argmin()][0])
        extBot = tuple(c[c[:, :, 1].argmax()][0])
        new_image = image[extTop[1]:extBot[1], extLeft[0]:extRight[0]]
        if plot:
            plt.figure(figsize=(8,4))
            plt.subplot(1,2,1)
            plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
            plt.title("Original Image")
            plt.axis('off')
            plt.subplot(1,2,2)
            plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_BGR2RGB))
            plt.title("Cropped Image")
            plt.axis('off')
            plt.show()
        return new_image
    else:
        return image

# Crop and resize training images to 299x299
for cls in classes:
    src_dir = os.path.join(original_train_dir, cls)
    dst_dir = os.path.join(base_train_dir, cls)
    if not os.path.exists(src_dir):
        print(f"Warning: Training class folder '{src_dir}' missing, skipping.")
        continue
    for j, img_name in enumerate(tqdm(os.listdir(src_dir), desc=f"Processing Training class {cls}")):
        img_path = os.path.join(src_dir, img_name)
        img = cv2.imread(img_path)
        if img is not None:
            cropped = crop_image(img)
            resized = cv2.resize(cropped, (299, 299))
            cv2.imwrite(os.path.join(dst_dir, f"{j}.jpg"), resized)

# Crop and resize testing images similarly
for cls in classes:
    src_dir = os.path.join(original_test_dir, cls)
    dst_dir = os.path.join(base_test_dir, cls)
    if not os.path.exists(src_dir):
        print(f"Warning: Testing class folder '{src_dir}' missing, skipping.")
        continue
    for j, img_name in enumerate(tqdm(os.listdir(src_dir), desc=f"Processing Testing class {cls}")):
        img_path = os.path.join(src_dir, img_name)
        img = cv2.imread(img_path)
        if img is not None:
            cropped = crop_image(img)
            resized = cv2.resize(cropped, (299, 299))
            cv2.imwrite(os.path.join(dst_dir, f"{j}.jpg"), resized)

# Data augmentation and generators
datagen = ImageDataGenerator(rotation_range=10,
                             height_shift_range=0.2,
                             horizontal_flip=True,
                             validation_split=0.2)

train_generator = datagen.flow_from_directory(base_train_dir,
                                              target_size=(299, 299),
                                              batch_size=32,
                                              class_mode='categorical',
                                              subset='training',
                                              shuffle=True)

valid_generator = datagen.flow_from_directory(base_train_dir,
                                              target_size=(299, 299),
                                              batch_size=32,
                                              class_mode='categorical',
                                              subset='validation',
                                              shuffle=True)

test_datagen = ImageDataGenerator()
test_generator = test_datagen.flow_from_directory(base_test_dir,
                                                  target_size=(299, 299),
                                                  batch_size=32,
                                                  class_mode='categorical',
                                                  shuffle=False)

# Build InceptionV3 based model
base_model = InceptionV3(weights='imagenet', include_top=False, input_shape=(299, 299, 3))
x = GlobalAveragePooling2D()(base_model.output)
x = Dropout(0.5)(x)
output = Dense(len(classes), activation='softmax')(x)
model = Model(inputs=base_model.input, outputs=output)

model.compile(optimizer=Adam(learning_rate=0.0001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Callbacks
checkpoint = ModelCheckpoint('best_model_inceptionv3.h5', monitor='val_accuracy', save_best_only=True, verbose=1)
earlystop = EarlyStopping(monitor='val_loss', patience=10, verbose=1, restore_best_weights=True)

# Train the model
history = model.fit(train_generator,
                    validation_data=valid_generator,
                    epochs=30,
                    callbacks=[checkpoint, earlystop])

# Plot training history
epochs = range(1, len(history.history['accuracy']) + 1)
plt.figure(figsize=(10, 6))
plt.plot(epochs, history.history['accuracy'], 'go-', label='Train Accuracy')
plt.plot(epochs, history.history['val_accuracy'], 'ro-', label='Validation Accuracy')
plt.plot(epochs, history.history['loss'], 'g^-', label='Train Loss')
plt.plot(epochs, history.history['val_loss'], 'r^-', label='Validation Loss')
plt.title('InceptionV3 Model Accuracy and Loss')
plt.xlabel('Epochs')
plt.ylabel('Value')
plt.legend()
plt.show()

# Predict and visualize random test samples
class_labels = {v: k for k, v in train_generator.class_indices.items()}
image_paths = []
for cls in classes:
    cls_dir = os.path.join(base_test_dir, cls)
    if os.path.exists(cls_dir):
        image_paths.extend([os.path.join(cls_dir, f) for f in os.listdir(cls_dir) if f.endswith(('.jpg', '.png'))])
random_samples = random.sample(image_paths, min(len(image_paths), 10))

plt.figure(figsize=(20, 8))
for i, img_path in enumerate(random_samples):
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_resized = cv2.resize(img_rgb, (299, 299))
    img_expanded = np.expand_dims(img_resized, axis=0) / 255.0

    preds = model.predict(img_expanded)
    pred_class = class_labels[np.argmax(preds)]

    plt.subplot(2, 5, i + 1)
    plt.imshow(img_rgb)
    plt.title(f"Predicted: {pred_class}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score, recall_score


# Get true labels and predicted probabilities from test generator
test_steps = test_generator.samples // test_generator.batch_size
if test_generator.samples % test_generator.batch_size != 0:
    test_steps += 1

y_true = test_generator.classes
y_pred_probs = model.predict(test_generator, steps=test_steps, verbose=1)
y_pred = np.argmax(y_pred_probs, axis=1)

# Accuracy
accuracy = accuracy_score(y_true, y_pred)

# F1 Score (macro and per class)
f1_macro = f1_score(y_true, y_pred, average='macro')
f1_per_class = f1_score(y_true, y_pred, average=None)

# Sensitivity (Recall)
recall_macro = recall_score(y_true, y_pred, average='macro')
recall_per_class = recall_score(y_true, y_pred, average=None)

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)

print("Accuracy:", accuracy)
print("Macro F1 Score:", f1_macro)
print("Per Class F1 Score:", f1_per_class)
print("Macro Sensitivity (Recall):", recall_macro)
print("Per Class Sensitivity (Recall):", recall_per_class)
print("Confusion Matrix:\n", cm)

# Optionally, display a classification report with precision, recall, f1 per class
print("\nClassification Report:\n", classification_report(y_true, y_pred, target_names=classes))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report

def plot_classification_report_extended(y_true, y_pred, classes, figsize=(12, 6)):
    report = classification_report(y_true, y_pred, target_names=classes, output_dict=True)

    # Prepare rows for per-class, macro avg, weighted avg
    rows = classes + ['macro avg', 'weighted avg']
    metrics = ['precision', 'recall', 'f1-score']
    data = []
    for row in rows:
        row_data = [report[row][metric] if metric in report[row] else 0 for metric in metrics]
        data.append(row_data)

    df = pd.DataFrame(data, index=rows, columns=metrics)

    # Colors inspired by research standards (Tableau 10 palette samples)
    colors = ['#1f77b4',  # muted blue
              '#ff7f0e',  # safety orange
              '#2ca02c']  # cooked asparagus green

    ax = df.plot(kind='bar', figsize=figsize, color=colors)
    ax.set_title('Classification Report: Per Class, Macro & Weighted Avg')
    ax.set_ylabel('Score')
    ax.set_ylim(0, 1.05)
    ax.set_xlabel('Classes')
    ax.set_xticks(np.arange(len(rows)))
    ax.set_xticklabels(rows, rotation=45)
    ax.legend(loc='lower right')
    plt.tight_layout()
    plt.show()

# Example usage:
plot_classification_report_extended(y_true, y_pred, classes)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.metrics import classification_report

def plot_classification_report(y_true, y_pred, classes, figsize=(8,6), cmap='Blues'):
    report = classification_report(y_true, y_pred, target_names=classes, output_dict=True)
    # Remove accuracy and averages from the dict for plotting
    classes_metrics = {k: v for k, v in report.items() if k in classes}

    metrics = ['precision', 'recall', 'f1-score', 'support']
    values = np.array([[metrics_info[m] for m in metrics] for cls, metrics_info in classes_metrics.items()])
    # Plot support separately for better scale by normalizing or use separate axis if needed

    fig, ax = plt.subplots(figsize=figsize)
    sns.heatmap(values[:, :-1], annot=True, fmt=".2f", cmap=cmap, cbar=True,
                yticklabels=classes, xticklabels=metrics[:-1], ax=ax)
    ax.set_title("Classification Report Metrics")
    plt.show()

# Use in your code after predictions like:
plot_classification_report(y_true, y_pred, classes)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from itertools import cycle
import numpy as np

# Binarize the true labels for multiclass ROC (one-hot encoding)
y_true_bin = label_binarize(y_true, classes=range(len(classes)))
n_classes = y_true_bin.shape[1]

# Get prediction probabilities for each class
y_score = y_pred_probs

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Compute micro-average ROC curve and ROC area
fpr["micro"], tpr["micro"], _ = roc_curve(y_true_bin.ravel(), y_score.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# Plot all ROC curves
plt.figure(figsize=(10, 8))
colors = cycle(['aqua', 'darkorange', 'cornflowerblue', 'green'])  # Use research paper style colors

for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label=f'ROC curve of class {classes[i]} (area = {roc_auc[i]:0.2f})')

plt.plot(fpr["micro"], tpr["micro"],
         label='micro-average ROC curve (area = {0:0.2f})'.format(roc_auc["micro"]),
         color='deeppink', linestyle=':', linewidth=3)

plt.plot([0, 1], [0, 1], 'k--', lw=2)  # Diagonal line for random guess
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multiclass ROC Curves')
plt.legend(loc="lower right")
plt.show()
```
